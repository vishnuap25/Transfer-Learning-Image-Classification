### Import Libraries

```python
from tensorflow.keras.preprocessing import image
import tensorflow as tf
import os
import numpy as np
```

### Setting input Image Size

```python
IMAGE_SIZE = [224, 224]
```

### Getting Output classes from training Directory to label the inference

```python
output_classes = os.listdir('data/training_set/')
output_classes
```

### Loading Trained Model

```python
model = tf.keras.models.load_model('model/classification_model.h5')
```

### Loading Image to be inferenced

```python
img = image.load_img('1969-Ford-Mustang-Boss-429-7.jpg', target_size=IMAGE_SIZE)
```

```python
img
```

### Convering the Image to arrays

```python
img_array = image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
```

### Getting the inference from the model

```python
predictions = model.predict(img_array)
score = predictions[0]
```

```python
predictions
```

```python
result = np.where(predictions == 1)[1][0]
```

### Final Output

```python
output_classes[result]
```
